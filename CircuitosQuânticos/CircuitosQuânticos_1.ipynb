# Circuitos Quânticos 1
Códigos utilizados na Seção `Circuitos Quânticos I` dos slides.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.visualization import visualize_transition
from qiskit.visualization import plot_bloch_multivector
from numpy import pi
```

# Exemplo 1
$\ket{\psi} = X \ket{0}$

```python
qreg_q = QuantumRegister(1, 'q')
circuit = QuantumCircuit(qreg_q)

circuit.x(qreg_q[0])

circuit.draw("mpl", filename='circuito_exemplo1.png')
```

```python
plot_bloch_multivector(circuit)
```

# Exemplo 2
$\ket{\psi} = Z \cdot X \ket{0}$

```python
qreg_q = QuantumRegister(1, 'q')
circuit = QuantumCircuit(qreg_q)

circuit.x(qreg_q[0])
circuit.z(qreg_q[0])

circuit.draw("mpl", filename='circuito_exemplo2.png')
```

```python
plot_bloch_multivector(circuit)
```

# Exemplo 3
$\ket{\psi} = H \cdot Z \cdot X \ket{0}$

```python
qreg_q = QuantumRegister(1, 'q')
circuit = QuantumCircuit(qreg_q)

circuit.x(qreg_q[0])
circuit.z(qreg_q[0])
circuit.h(qreg_q[0])

circuit.draw("mpl", filename='circuito_exemplo3.png')
```

```python
plot_bloch_multivector(circuit)
```

# Exemplo 4

```python
qreg_q = QuantumRegister(2, 'q')
circuit = QuantumCircuit(qreg_q)

circuit.x(qreg_q[0])

circuit.draw("mpl", filename='circuito_exemplo4.png')
```

# Exemplo 6
$ \ket{\psi} = (H \otimes X \otimes I) \ket{011} $

ou

$ \ket{\psi} = H \ket{0} \otimes X\ket{1} \otimes I\ket{1} $

```python
qreg_q = QuantumRegister(3, 'q')
circuit = QuantumCircuit(qreg_q)

circuit.x(qreg_q[0])
circuit.h(qreg_q[2])

circuit.draw("mpl", filename='circuito_exemplo6.png')
```

```python
plot_bloch_multivector(circuit)
```

# Exemplo 7
$ \ket{\psi} = \bra{0} \cdot Z\ket{1} $

```python
qreg_q = QuantumRegister(1, 'q')
creg_c = ClassicalRegister(1, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.x(qreg_q[0])
circuit.z(qreg_q[0])
circuit.measure(qreg_q[0], creg_c[0]).c_if(creg_c, 0)

circuit.draw("mpl", filename='circuito_exemplo7.png')
```

# Exemplo 8
$ \ket{\psi} =  \ket{1} \bra{0} \cdot X \cdot \ket{0} $

```python
qreg_q = QuantumRegister(2, 'q')
creg_c = ClassicalRegister(1, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.x(qreg_q[0])
circuit.x(qreg_q[1]).c_if(creg_c, 1)

circuit.draw("mpl", filename='circuito_exemplo8.png')
```

```python
plot_bloch_multivector(circuit)
```
